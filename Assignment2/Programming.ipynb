```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from cvxpy import *
```

# Question 1, 2 and 4

```python
# getting the data
data = h5py.File("toy.hdf5",'r')
X, y = np.array(data.get('X')).T, np.array(data.get('y'))
```

```python
# Defining variables to optimize over, the parameters and the cost function
beta = Variable(X.shape[1])
beta_0 = Variable()
y_pos, y_neg = np.where(y==1), np.where(y==-1)
slack = Variable(X.shape[0])
C1 = Parameter(sign="positive")
C2 = Parameter(sign="positive")
cost = 0.5 * norm(beta)**2 + C1 * sum_entries(slack[y_pos[0]]) + C2 * sum_entries(slack[y_neg[0]])
```

```python
# Defining constraints
constraints = [slack >= 0]
for i in range(X.shape[0]):
    constraints.append((y[i]*(X[i]*beta+beta_0) >= 1-slack[i]))
```

```python
# Defining objective and the problem with both - objective and constraints
objective = Minimize(cost)
prob = Problem(objective, constraints)
```

```python
param_list = [(1, 1), (1,10) , (10,1)]
i = 1
plt.figure()
for c1, c2 in param_list:
    # Setting value of parameters and finding optimal value there
    C1.value = c1
    C2.value = c2
    print("Optimal value (C1 = %f, C2 = %f): %f" % (c1, c2, prob.solve()))

    labels = [0]*X.shape[0]
    sl = np.array(slack.value)
    correct = []
    incorrect = []
    weighted_err = 0
    for j in range(X.shape[0]):
        if sl[j][0] < 1e-5:
            correct.append(X[j])
        else:
            if y[j] == 1:
                weighted_err += c1
            else:
                weighted_err += c2
            incorrect.append(X[j])
    print("Weighted error : %d" % weighted_err)
    correct = np.array(correct)
    incorrect = np.array(incorrect)
    #plt.subplot(3,1,i)
    #Plot the points
    plt.scatter(correct[:,0], correct[:,1], marker='o')
    plt.scatter(incorrect[:,0], incorrect[:,1], marker='^')
    
    #Plot the decision boundary
    plt.plot(list(map(lambda x: -float(beta_0.value+float(beta.value[0])*x)/float(beta.value[1]), range(-5,14,1))),range(-5,14,1), '-r')
    
    #Plot the margins
    plt.plot(list(map(lambda x: -float(beta_0.value+float(beta.value[0])*x-1)/float(beta.value[1]), range(-5,14,1))),range(-5,14,1), '--b')
    plt.plot(list(map(lambda x: -float(beta_0.value+float(beta.value[0])*x+1)/float(beta.value[1]), range(-5,14,1))),range(-5,14,1), '--g')
    i+=1
    plt.show()
```

In the above the points with marker '^' are misclassified.

Due to numerical inconsistensies, I consider every value smaller than $10^{-5}$ as 0

With different penalty parameters, the boundary decision changes since it tries to minimize the errors which means trying to have a better classification for the class with higher penalty

In the second one, it moves a bit up to reduce misclassification for that class and as expected, in the third one, it moves a bit down since we have the opposite penalties
